```python
# first things first:
# check no 2 nodes with the same lines in the same group
```

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/Hubs_groups_with_full_data_flat.csv', encoding='windows-1255')
```

```python
df.head(3)
```

```python
df.set_index('Column1', inplace=True)
```

```python
suspected_nodes = {g:[] for g in df['group'].unique()}
```

```python
for group, nodes in df.groupby('group')['node']:
  node_list = nodes.tolist()
  print(f'Group: {group}, Nodes: {node_list}')
  if len(node_list) == 1:
    continue
  for i, node_a in enumerate(node_list):
    for node_b in node_list[i+1:]:
      if set(df[df['node']==node_a]['Line_Unique'].to_list()) == set(df[df['node']==node_b]['Line_Unique'].to_list()):
        suspected_nodes[group].append(node_a)
        suspected_nodes[group].append(node_b)
```

```python
for k in list(suspected_nodes.keys()):
  if len(suspected_nodes[k]) == 0:
    suspected_nodes.pop(k)
```

```python
for k in suspected_nodes.keys():
  suspected_nodes[k] = set(suspected_nodes[k])
```
